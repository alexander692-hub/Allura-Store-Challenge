### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#suma la venta total en cada tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

#imprime el total de cada tienda
print("Ingreso total tienda 1:", ingreso_tienda1)
print("Ingreso total tienda 2:", ingreso_tienda2)
print("Ingreso total tienda 3:", ingreso_tienda3)
print("Ingreso total tienda 4:", ingreso_tienda4)
```

```python
#imrpime el total vendido en las 4 tiendas

ingreso_total = 0

ingreso_total += tienda['Precio'].sum()
ingreso_total += tienda2['Precio'].sum()
ingreso_total += tienda3['Precio'].sum()
ingreso_total += tienda4['Precio'].sum()

print(f"El ingreso total de las 4 tiendas es: {ingreso_total}")
```

# 2. Ventas por categoría

```python
#analiza ventas de articulos por categoria en cada tienda
def ventas_por_categoria(tienda, nombre_tienda):
    ventas = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
    print(f"\n--- Tienda {nombre_tienda} ---")
    print(ventas)

#imprime ventas de articulos cada tienda
ventas_por_categoria(tienda, "1")
ventas_por_categoria(tienda2, "2")
ventas_por_categoria(tienda3, "3")
ventas_por_categoria(tienda4, "4")
```

# 3. Calificación promedio de la tienda


```python
#analiza la valoracion en cada tienda
def valoracion(tienda, nombre_tienda):
    media = tienda.groupby('Calificación')['Calificación'].count().sort_values(ascending=False)
    print(f"\n--- Tienda {nombre_tienda} ---")
    print(media)

#imprime la valoracion de cada tienda
valoracion(tienda, "1")
valoracion(tienda2, "2")
valoracion(tienda3, "3")
valoracion(tienda4, "4")
```

# 4. Productos más y menos vendidos

```python

def productos_vendidos(tienda, nombre_tienda):
    ventas_producto = tienda.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

    print(f"\n--- Tienda {nombre_tienda} ---")
    print("Productos mas vendidos:")
    print(ventas_producto.head(5))  # Muestra los 5 mas vendidos
    print("\nProductos menos vendidos:")
    print(ventas_producto.tail(5))  # Muestra los 5 menos vendidos

# imprime los productos mas y menos vendidos de cada tienda
productos_vendidos(tienda, "1")
productos_vendidos(tienda2, "2")
productos_vendidos(tienda3, "3")
productos_vendidos(tienda4, "4")
```

# 5. Envío promedio por tienda

```python
# 5 Envio promedio por tienda

def costo_envio_promedio(tienda, nombre_tienda):
    costo_promedio = tienda['Costo de envío'].mean()
    # Redondear el costo promedio a 2 decimales
    costo_promedio_redondeado = round(costo_promedio, 2)
    print(f"\n--- Tienda {nombre_tienda} ---")
    print(f"Costo de envio: {costo_promedio_redondeado}")

costo_envio_promedio(tienda, "1")
costo_envio_promedio(tienda2, "2")
costo_envio_promedio(tienda3, "3")
costo_envio_promedio(tienda4, "4")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ingresos totales por tienda
ingresos = {
    'Tienda 1': ingreso_tienda1,
    'Tienda 2': ingreso_tienda2,
    'Tienda 3': ingreso_tienda3,
    'Tienda 4': ingreso_tienda4
}

tiendas = list(ingresos.keys())
valores = list(ingresos.values())

plt.figure(figsize=(8, 6))
barras = plt.bar(tiendas, valores, color=['skyblue', 'lightcoral', 'lightgreen', 'red'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos')

# Agregar el total de ventas arriba de cada barra
for barra, valor in zip(barras, valores):
    plt.text(barra.get_x() + barra.get_width() / 2, barra.get_height() + 10,
             round(valor, 2), ha='center', va='bottom')

plt.show()


# Distribucion de categorias de productos para las 4 tiendas
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

for i, (tienda_actual, nombre_tienda) in enumerate(zip(tiendas, nombres_tiendas)):
    fila = i // 2
    columna = i % 2
    ax = axes[fila, columna]

    categorias = tienda_actual['Categoría del Producto'].value_counts()
    ax.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Distribución de Categorías ({nombre_tienda})')
    ax.axis('equal')

plt.tight_layout()
plt.show()


# Costo de envio promedio por tienda
costos_envio = {
    'Tienda 1': tienda['Costo de envío'].mean(),
    'Tienda 2': tienda2['Costo de envío'].mean(),
    'Tienda 3': tienda3['Costo de envío'].mean(),
    'Tienda 4': tienda4['Costo de envío'].mean()
}

tiendas_envio = list(costos_envio.keys())
costos_promedio = list(costos_envio.values())

plt.figure(figsize=(8, 6))
sns.lineplot(x=tiendas_envio, y=costos_promedio, marker='o', color='mediumpurple')
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo Promedio de Envío')
plt.show()
```
